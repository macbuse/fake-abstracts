```python
import pandas as pd
import pickle
import arxivscraper as ax

from datetime import timedelta, date
import time


start_day= date(2018, 1, 1)
end_day = date(2018, 12, 31)


week, data = start_day, []

while week <= end_day:
    print('\n')
    weekend = week + timedelta(days=6) 
    if weekend > end_day: weekend = end_day
    scraper = ax.Scraper(category='math',
                         date_from= str(week), 
                         date_until=str(weekend),
                         t=10)
    output = scraper.scrape()
    data.extend(output)
    week += timedelta(days=7) 

    
    time.sleep(10)
    
```

# Dump

dump the data somewhere for future use.
FYI it's a list of dictionnaries

```python
import pickle

fn = 'arx_math_%s_%s.pkl' %(str(start_day), str(end_day))
with open(fn ,'wb') as fp:
    pickle.dump(data,fp)
```
